## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Create copies (safe practice)
users_df = users.copy()
flights_df = flights.copy()
hotels_df = hotels.copy()

# Convert date columns (ONLY where present)
flights_df['date'] = pd.to_datetime(flights_df['date'])
hotels_df['date'] = pd.to_datetime(hotels_df['date'])

# Check again for safety
print("Users Missing:\n", users_df.isnull().sum())
print("\nFlights Missing:\n", flights_df.isnull().sum())
print("\nHotels Missing:\n", hotels_df.isnull().sum())

# No missing values → no imputation needed
# No duplicates already → no removal needed

# Feature Engineering on flights dataset (correct dataset)

# Flight duration category
flights_df['time_category'] = pd.cut(
    flights_df['time'],
    bins=[0, 1, 2, 3],
    labels=['short', 'medium', 'long']
)

# Distance category
flights_df['distance_category'] = pd.cut(
    flights_df['distance'],
    bins=[0, 400, 700, 1000],
    labels=['short', 'medium', 'long']
)

# Price category (for segmentation understanding)
flights_df['price_category'] = pd.qcut(
    flights_df['price'],
    3,
    labels=['low', 'medium', 'high']
)

# Route feature (EDA purpose)
flights_df['route'] = flights_df['from'] + " → " + flights_df['to']

# Price efficiency
flights_df['price_per_km'] = flights_df['price'] / flights_df['distance']

# Extract date features
flights_df['year'] = flights_df['date'].dt.year
flights_df['month'] = flights_df['date'].dt.month
flights_df['day'] = flights_df['date'].dt.day

# Final check
flights_df.head()
````

````python
# Chart - 33: High Distance but Low Price (Underpricing)

route_stats = flights.groupby(['from','to']).agg({'distance':'mean','price':'mean'}).reset_index()
route_stats['route'] = route_stats['from'] + " → " + route_stats['to']

# filter: high distance but low price
filtered = route_stats[(route_stats['distance'] > route_stats['distance'].median()) &
                       (route_stats['price'] < route_stats['price'].median())]

top_underpriced = filtered.sort_values(by='distance', ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(data=top_underpriced, x='route', y='distance', palette='Blues')
plt.title("High Distance but Low Price Routes (Underpricing)")
plt.xticks(rotation=45)
plt.show()
````
